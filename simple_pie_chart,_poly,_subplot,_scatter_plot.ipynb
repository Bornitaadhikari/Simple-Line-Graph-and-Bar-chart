```python
import matplotlib.pyplot as plt
import numpy as np                            
```

#subplot




```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([4,2,8,10])
plt.subplot(1,2,1)

plt.plot(x,y,color='red',linestyle='--',linewidth=3,marker='*')
plt.title('Sports Watch data')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.grid()

#plot 2
x=np.array([0,1,2,3])
y=np.array([15,25,35,40])
plt.subplot(1,2,2)

plt.plot(x,y,color='yellow',linestyle='--',linewidth=3,marker='*')
plt.title('Sports Watch data')
plt.xlabel('Average Pulse')
plt.grid()
plt.show()
```

#Scatter


```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.figure(figsize=(5,3),dpi=200)
plt.scatter(x, y,c=colors,cmap='Accent')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y,c=colors,s=sizes, cmap='Accent')


plt.colorbar()

plt.show()
```

#Pie Chart

```python
y=([20,35,10,35])
mycolors=['Apple','Orange','Banana','Mango']
mycolors=['black','blue','yellow','green']
plt.figure(figsize=(5,3),dpi=200)
plt.pie(y,colors=mycolors,labels=mycolors)
plt.legend(title='FRUITS')
plt.show()
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```
