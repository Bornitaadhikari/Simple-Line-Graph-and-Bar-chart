```python
import matplotlib.pyplot as plt
from matplotlib import rc

import matplotlib.font_manager
import os
import pandas as pd
import numpy as np
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#Resize the graph
plt.figure(figsize=(5,3),dpi=150)
#plt.plot(x,y,color='red',linestyle='--',linewidth=3,marker='.',markersize=8,markeredgecolor='blue',label='2x')
plt.plot(x,y,'b^--',label='2x')
#line number two
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,color='red',linewidth=5)
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title("Our First Graph!",fontdict={'fontname':'Garamond','fontsize':30})
plt.xlabel("X Axsis")
plt.ylabel("Y Axsis")
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.savefig('mygraph.png')
plt.legend()

plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
patterns=['/','o','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(10,10))
plt.show()  
```
